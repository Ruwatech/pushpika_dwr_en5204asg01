```python
import cv2 as cv
import matplotlib.pyplot as plt
im = cv.imread('tim1.jpg',cv.IMREAD_COLOR)
assert im is not None
imrgb = cv.cvtColor(im, cv.COLOR_BGR2RGB)
plt.imshow(imrgb)
plt.show()
```

```python
blurred = cv.blur(imrgb,(5,5))
gausian_blurred = cv.GaussianBlur(imrgb,(21,21),4)
fig,ax = plt.subplots(1,3,figsize=(10,30))
ax[0].imshow(imrgb)
ax[1].imshow(blurred)
ax[2].imshow(gausian_blurred)
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
transform = np.arange(0, 256).astype('uint8')
transform[0] = 255
fig,ax = plt.subplots()

ax.plot(transform)

ax.set_xlabel( r'Input, $f(\mathbf{x})$')
ax.set_ylabel('Output, $\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.set_aspect('equal')
plt.savefig('transform.png')
plt.show()
img_orig = cv.imread('lung.jpg', cv.IMREAD_GRAYSCALE)
jake = img_orig
print(img_orig.shape)
fig,ax = plt.subplots(1,2,figsize=(100,300))
ax[0].imshow(jake,cmap='gray', vmin=0, vmax=255)


image_transformed = cv.LUT(img_orig, transform)
ax[1].imshow(image_transformed,cmap='gray', vmin=0, vmax=255)
#ax[1].imshow(image_transformed)
plt.show()
```

```python


# Load the image
image = cv.imread("tim1.jpg")

# Create LUT array with invert function
lut = np.arange(256)[::-1]

# Apply LUT to image
result = cv.LUT(image, lut)

# Display the original image and the transformed image
cv.imshow("Original", image)
cv.imshow("Transformed", result)
cv.waitKey(0)
cv.destroyAllWindows()
```
